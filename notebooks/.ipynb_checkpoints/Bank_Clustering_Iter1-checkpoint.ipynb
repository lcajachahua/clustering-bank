# Bank Dataset Clustering Optimal K Search

Using NBClust Library

Instalar previamente la librería ejecutando en la Terminal:
conda install -c conda-forge r-nbclust

```r
library(NbClust)
```

```r
data<-read.csv("../data/segmentacion.csv")
```

```r
mat<-data[,2:25]
```

```r
head(data)
```

```r
NbClust(mat, diss=NULL, distance="maximum", min.nc=4, max.nc=12, method="median", index="all")
```

```r
nbclu<-NbClust(mat, diss=NULL, distance="euclidean", min.nc=4, max.nc=12, method="kmeans", index="all")
```

```r
hist(nbclu$Best.partition)
```

```r
data<-cbind(data,'cluster'=nbclu$Best.partition)
```

```r
data$cluster<-as.factor(data$cluster)
```
